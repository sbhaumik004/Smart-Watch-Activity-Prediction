#9 Bifurcate the data as per predictors and target

```python
x= Data.drop('activity',axis=1)
y= Data.activity
```

#10. Bifurcate the data for training and testing the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,random_state=150,test_size=0.8)
```

```python
x_train
```

```python
y_train
```

#11. Train the data with appropriate model.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=255,random_state=20)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random=RandomizedSearchCV(RandomForestClassifier(),{"n_estimators":[150,200,255],"random_state":[20,30,40]},n_iter=12,verbose=2)
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

#12. Find out the accuracy which serves the best.

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score,precision_score
```

```python
accuracy_score(y_test,model.predict(x_test))
```

#13. Create a confusion matrix for the results achieved.

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
sns.heatmap(confusion_matrix(y_test, model.predict(x_test)), annot=True);
```

#14.Find out the Precision and Recall score.

```python
print(classification_report(y_test,model.predict(x_test)))
```
